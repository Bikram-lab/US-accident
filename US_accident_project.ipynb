```python
# Convert datetime columns with mixed format handling
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='mixed', errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], format='mixed', errors='coerce')

# Extract time components
df['Hour'] = df['Start_Time'].dt.hour
df['Day_of_Week'] = df['Start_Time'].dt.dayofweek  # 0=Monday, 6=Sunday
df['Month'] = df['Start_Time'].dt.month
df['Year'] = df['Start_Time'].dt.year
df['Day_of_Month'] = df['Start_Time'].dt.day

# Check for any null values
print("\nNull values in temporal features:")
print(df[['Hour', 'Day_of_Week', 'Month', 'Year']].isnull().sum())
```

```python
# Rush hour (7-9, 16-19)
df['Is_Rush_Hour'] = ((df['Hour'] >= 7) & (df['Hour'] <= 9)) | \
                     ((df['Hour'] >= 16) & (df['Hour'] <= 19))
df['Is_Weekend'] = df['Day_of_Week'].isin([5, 6])

print("Rush Hour distribution:")
print(df['Is_Rush_Hour'].value_counts())

print("\nWeekend distribution:")
print(df['Is_Weekend'].value_counts())
```

```python
# Accident Duration
df['Duration_Minutes'] = (df['End_Time'] - df['Start_Time']).dt.total_seconds() / 60
df['Duration_Minutes'] = df['Duration_Minutes'].clip(lower=0, upper=1440)  # Cap at 24 hours

print("Accident Duration Statistics:")
print(df['Duration_Minutes'].describe())
```

```python
# Time of day categories
def categorize_time(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Time_of_Day'] = df['Hour'].apply(categorize_time)

print("Time of Day distribution:")
print(df['Time_of_Day'].value_counts())
```

```python
# Fill missing weather values
df['Temperature(F)'].fillna(df['Temperature(F)'].median(), inplace=True)
df['Humidity(%)'].fillna(df['Humidity(%)'].median(), inplace=True)
df['Pressure(in)'].fillna(df['Pressure(in)'].median(), inplace=True)
df['Visibility(mi)'].fillna(df['Visibility(mi)'].median(), inplace=True)
df['Wind_Speed(mph)'].fillna(0, inplace=True)
df['Precipitation(in)'].fillna(0, inplace=True)

print("\nWeather features summary:")
print(df[['Temperature(F)', 'Humidity(%)', 'Visibility(mi)', 'Wind_Speed(mph)']].describe())
```

```python
# Weather risk indicators
df['Low_Visibility'] = (df['Visibility(mi)'] < 5).astype(int)
df['High_Wind'] = (df['Wind_Speed(mph)'] > 20).astype(int)
df['Has_Precipitation'] = (df['Precipitation(in)'] > 0).astype(int)

# Composite weather risk score
df['Weather_Risk_Score'] = (
    df['Low_Visibility'] +
    df['High_Wind'] +
    df['Has_Precipitation']
)
print("\nWeather Risk Score distribution:")
print(df['Weather_Risk_Score'].value_counts().sort_index())
```

```python
# Simplified weather condition
df['Weather_Condition'].fillna('Clear', inplace=True)

def simplify_weather(condition):
    condition = str(condition).lower()
    if any(x in condition for x in ['rain', 'drizzle', 'shower']):
        return 'Rain'
    elif any(x in condition for x in ['snow', 'sleet', 'ice']):
        return 'Snow'
    elif any(x in condition for x in ['fog', 'mist', 'haze']):
        return 'Fog'
    elif any(x in condition for x in ['cloud', 'overcast']):
        return 'Cloudy'
    elif any(x in condition for x in ['clear', 'fair']):
        return 'Clear'
    else:
        return 'Other'

df['Weather_Simple'] = df['Weather_Condition'].apply(simplify_weather)

print("Simplified Weather Conditions:")
print(df['Weather_Simple'].value_counts())
```

```python
# Day/Night based on Sunrise_Sunset
df['Is_Night'] = (df['Sunrise_Sunset'] == 'Night').astype(int)

print("Day/Night distribution:")
print(df['Is_Night'].value_counts())
```

```python
# Define critical features to keep
critical_features = [
    # Target
    'Severity',

    # Temporal
    'Hour', 'Day_of_Week', 'Month', 'Year',
    'Is_Rush_Hour', 'Is_Weekend', 'Time_of_Day',
    'Duration_Minutes',

    # Traffic Control (Original)
    'Junction', 'Traffic_Signal', 'Stop', 'Crossing',
    'Railway', 'Station', 'Amenity',

    # Traffic Control (Engineered)
    'Traffic_Control_Count', 'Has_Traffic_Control',
    'Has_Signal_Control', 'Has_Intersection',

    # Weather
    'Temperature(F)', 'Humidity(%)', 'Pressure(in)',
    'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)',
    'Low_Visibility', 'High_Wind', 'Has_Precipitation',
    'Weather_Risk_Score', 'Weather_Simple',

    # Location/Time
    'Distance(mi)', 'Is_Night', 'State'
]

# Create final processed dataset
df_processed = df[critical_features].copy()

print(f"\nOriginal features: {len(df.columns)}")
print(f"Final features: {len(df_processed.columns)}")
print(f"Records: {len(df_processed):,}")
print(f"\nFinal Dataset Shape: {df_processed.shape}")
```
